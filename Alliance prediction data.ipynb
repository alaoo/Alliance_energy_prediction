```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/energydata_complete.csv')
data.head(5)
```

```python
data.shape
```

```python
target = data['Appliances']
features = data.drop(['Appliances', 'date', 'lights'], axis=1)
features.head(2)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features, target, test_size=0.3, random_state=42)

Scalar=MinMaxScaler()
Scalar.fit(X_train)
X_train_scale=Scalar.transform(X_train)
X_test_scale=Scalar.transform(X_test)
X_train=pd.DataFrame(X_train_scale, columns=X_train.columns)
X_test=pd.DataFrame(X_test_scale, columns=X_test.columns)


model_1=LinearRegression()
model_1.fit(X_train, y_train)
y_hat=model_1.predict(X_test)

residual=y_test-y_hat
square_residual=np.square(residual)
rss=np.sum(square_residual)
R2=r2_score(y_test, y_hat)
MAE=mean_absolute_error(y_test, y_hat)

RMSE= np.sqrt(mean_squared_error(y_test, y_hat))

print(round(rss, 2))
```

```python
print(round(R2, 2))
```

```python
print(round(MAE, 2))
```

```python
print(round(RMSE, 3))
```

```python
model_2=Ridge(alpha=0.4)
model_2.fit(X_train, y_train)
y_hat=model_2.predict(X_test)

residual=y_test-y_hat
square_residual=np.square(residual)
rss=np.sum(square_residual)
R2=r2_score(y_test, y_hat)
MAE=mean_absolute_error(y_test, y_hat)

RMSE= np.sqrt(mean_squared_error(y_test, y_hat))

print(round(rss, 2))
print(round(R2, 2))
print(round(MAE, 2))
print(round(RMSE, 3))
```

```python
model_3=Lasso(alpha=0.001)
model_3.fit(X_train, y_train)
y_hat=model_3.predict(X_test)

residual=y_test-y_hat
square_residual=np.square(residual)
rss=np.sum(square_residual)
R2=r2_score(y_test, y_hat)
MAE=mean_absolute_error(y_test, y_hat)

RMSE= np.sqrt(mean_squared_error(y_test, y_hat))

print(round(rss, 2))
print(round(R2, 2))
print(round(MAE, 2))
print(round(RMSE, 3))
```
